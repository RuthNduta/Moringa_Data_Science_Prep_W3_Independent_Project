**Currently MTN Cote d'Ivoire would like to upgrade its technology infrastructure for its mobile users in Ivory Coast. Studying the given dataset, how does MTN Cote d'Ivoire go about the upgrade of its infrastructure strategy within the given cities?**

```
# Let's import the pandas and numpy libraries
# 
import pandas as pd
import numpy as np
```

```
# Let's first begin with loading datasets 
# We have 6 diffferent datasets that we need to analyse and draw conclusions on how MTN can go about upgrading its infrastructure strategy

#cells_geo_description.xlsx [Link] (Links to an external site.)
#cells_geo.csv [Link] (Links to an external site.)
#CDR_description.xlsx [Link] (Links to an external site.)
#CDR 20120507 [http://bit.ly/TelecomDataset1] (Links to an external site.)
#CDR 20120508 [http://bit.ly/TelecomDataset2] (Links to an external site.)
#CDR 20120509 [http://bit.ly/TelecomDataset3]
```

```
# loading the geo_csv dataset 
geo_df=pd.read_csv('cells_geo.csv')
geo_df = pd.read_csv("/content/cells_geo.csv", delimiter=";")
#renaming the geo columns
geo_df.rename(columns = {'SITE_CODE':'SITE_ID'}, inplace = True) 
geo_df
```

```
#loading the first Telecom dataset 
#renaming the columns for uniformity
col_names = ['PRODUCT',	'VALUE',	'DATE_TIME',	'CELL_ON_SITE',	'DW_A_NUMBER',	'DW_B_NUMBER',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID']
TelecomDataset1 = pd.read_csv('Telcom_dataset.csv',header=0, names=col_names)
TelecomDataset1
```

```
#loading the second Telecom dataset 
#renaming the columns for uniformity
col_names = ['PRODUCT',	'VALUE',	'DATE_TIME',	'CELL_ON_SITE',	'DW_A_NUMBER',	'DW_B_NUMBER',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID']
TelecomDataset2 = pd.read_csv('Telcom_dataset2.csv',header=0, names=col_names)
TelecomDataset2
```

```
#loading the third Telecom dataset 
#renaming the columns for uniformity
col_names = ['PRODUCT',	'VALUE',	'DATE_TIME',	'CELL_ON_SITE',	'DW_A_NUMBER',	'DW_B_NUMBER',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID']
TelecomDataset3 = pd.read_csv('Telcom_dataset3.csv',header=0, names=col_names)
TelecomDataset3
```

```
#concatenating the 3 telecom datasets together

df1_2_3= pd.concat([TelecomDataset1, TelecomDataset2, TelecomDataset3])
df1_2_3
```

```
#Joinig the 3datasets

df_all = df1_2_3.merge(geo_df, left_on="SITE_ID", right_on ="SITE_ID")
df_all
```

```
#Checking for any missing values in the merged data set

df_all.isnull().sum()
```

```
#There are 1997 missing values on Site Id column
# We fill the missing values with 0

df_all.fillna(value='0', inplace=True)
```

```
#Checking for missing values
df_all.isnull().sum()
```

```
df_all.info()
```

```
#Checking for duplicates

df_all.duplicated()
```

```
#dropping the duplicated data
df_all.drop_duplicates()
```

```
#Looking at top n records
df_all.head(5)
```

```
#Renaming the Value column to Billing price
df_all.rename(columns={'VALUE':'BILLING_PRICE'}, inplace=True)
```

```
#Describe the quick stats (count, mean, std, min, first quartile,median, third quartile, max) on THE PRICE columns

df_all[['BILLING_PRICE']].describe()
```

```
#dropping the unnamed column
df_all.drop
del df_all['Unnamed: 0']
df_all
```

```
#Cities with the most traffic

A = df_all['BILLING_PRICE'].max()
df_all.loc[df_all.BILLING_PRICE == 6750]
```

```
# Most popular product between voice and sms; we'll need TO convert the categorical data into integers

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Fitting the label encoder method in the column
label_encoder.fit(df_all['PRODUCT'])

#Converting the categories into integers
df_all['PRODUCT'] = label_encoder.transform(df_all['PRODUCT'])
#view the dataframe with the converted column
df_all

df_all['PRODUCT'].value_counts()

```
